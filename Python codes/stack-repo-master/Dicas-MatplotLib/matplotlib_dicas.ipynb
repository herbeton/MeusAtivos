# 2 - subplots

Existem três maneiras de gerar subplots:
- subplot()
- subplots()
- subplot2grid()

### 2.1 - subplot()


- Recebe como Argumentos :
    - Número de linhas
    - Número de colunas
    - index (índice)
- Retorna:
    - Objeto do tipo eixo que será utilizado para realizar o plot 

```python
vendas_ano_anterior = [37,39,40,45,40,42,48,40,43,46,45,48]
```

```python
fig=plt.figure(figsize=(14,6))

ax1=plt.subplot(1,2,1)
ax1.plot(meses, vendas)
ax1.set_title('Ano atual')
ax1.set_xlabel('Meses')
ax1.set_ylabel('Vendas [R$]')

ax2=plt.subplot(1,2,2)
ax2.plot(meses,vendas_ano_anterior)
ax2.set_title('Ano anterior')
ax2.set_xlabel('Meses')
ax2.set_ylabel('Vendas [R$]')

plt.show()
```

### 2.2 - subplots()


- Recebe como Argumentos :
    - Número de linhas
    - Número de colunas
- Retorna:
    - Array de eixos 

```python
fig,axes = plt.subplots(2,2)
print(axes)
#plt.tight_layout()
```

```python
fig,axes=plt.subplots(2,2,figsize=(14,8))

axes[0][0].plot(meses,vendas)
axes[0][0].set_title('Ano atual - plot 1')
axes[0][0].set_xlabel('Meses')
axes[0][0].set_ylabel('Vendas [R$]')

axes[0][1].plot(meses,vendas_ano_anterior)
axes[0][1].set_title('Ano Anterior - plot 2')
axes[0][1].set_xlabel('Meses')
axes[0][1].set_ylabel('Vendas [R$]')

axes[1][0].plot(meses,vendas)
axes[1][0].set_title('Ano Atual - Plot 3')
axes[1][0].set_xlabel('Meses')
axes[1][0].set_ylabel('Vendas [R$]')

axes[1][1].plot(meses,vendas_ano_anterior)
axes[1][1].set_title('Ano Anterior - plot 4')
axes[1][1].set_xlabel('Meses')
axes[1][1].set_ylabel('Vendas [R$]')

plt.tight_layout()
plt.show()
```

### 2.3- subplot2grid()
- Plots com diferentes tamanhos

- Recebe como Argumentos :
    - Shape - número de linhas e colunas
    - local do plot - número de linhas e colunas da localização do eixo na grid
    - linhas para o eixo se estender para a direita.
    - Número de colunas para o eixo abranger parte inferior
- Retorna:
    - Retorna o objeto de eixos para um local específico e de um tamanho específico em uma grid

```python
semestre_1 = meses[:6]
semestre_1
```

```python
fig=plt.figure(figsize=(8,8))

ax1=plt.subplot2grid(shape = (4,4),loc = (0,0),colspan=3)
ax1.plot(meses, vendas)
ax1.set_title('Plot 1 : (0,0)')

ax2=plt.subplot2grid((4,4),(0,3))
ax2.plot(semestre_1, vendas[:6])
ax2.set_title('Plot 2 : (0,3)')

ax3=plt.subplot2grid((4,4),(1,0),rowspan=3,colspan=3)
ax3.plot(meses,vendas_ano_anterior)
ax3.set_title('Plot 3 : (1,0)')

ax4=plt.subplot2grid((4,4),(1,3),rowspan=3,colspan=1)
ax4.plot(semestre_1,vendas_ano_anterior[:6])
ax4.set_title('Plot 4 : (1,3)')

fig.tight_layout()

plt.show()
```

# Bonus

### Gráfico de Radar

```python
meses = ['Jan', 'Fev', 'Mar','Abr','Mai','Jun','Jul','Ago','Set','Out','Nov','Dez']
vendas = [38,39,32,29,26,22,19,15,20,22,25,27]
vendas_ano_anterior = [37,39,40,45,40,42,48,40,43,46,45,48]
```

```python
labels = meses

angles = np.linspace(0, 2*np.pi, len(labels), endpoint=False)

# close the plot
vendas = np.concatenate((vendas,[vendas[0]]))
vendas2 = np.concatenate((vendas_ano_anterior,[vendas_ano_anterior[0]])) 
angles = np.concatenate((angles,[angles[0]]))

#Size of the figure
fig= plt.figure(figsize = (18,18))

ax = fig.add_subplot(221, polar= True)
ax.plot(angles, vendas, 'o-', linewidth=2, label = "2020", color= 'red')
ax.fill(angles, vendas, alpha=0.30, facecolor='red')
ax.set_thetagrids(angles * 180/np.pi, labels , fontsize = 13)


ax.set_rlabel_position(68)
plt.yticks([10,20,30,40,50], ["10",'20', "30", "40", "50"], color="gray", size=12)
plt.ylim(0,60)

ax.plot(angles, vendas2, 'o-', linewidth=2, label = "2019", color = 'green')
ax.fill(angles, vendas2, alpha=0.30, facecolor='green' )
ax.set_title('Vendas Loja 1')
ax.grid(True)

plt.legend(loc='best', bbox_to_anchor=(0.1, 0.1));
```
